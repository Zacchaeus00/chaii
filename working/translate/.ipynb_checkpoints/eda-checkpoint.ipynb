```python
squad1_translated = pd.read_csv('squad1.2_translated.csv')
squad1_translated
```

```python
for i in tqdm(range(len(squad1_translated))):
    if squad1_translated.loc[i, 'error'] == 'success':
        squad1_translated.loc[i, 'error_ta'] = 'success'
```

```python
squad1_translated.error_ta.isna().sum()
```

```python
squad1_translated.error_ta.value_counts()
```

```python
squad1_translated.drop(['error'], axis=1, inplace=True)
```

```python
squad1_translated.to_csv('squad1.2_translated.csv', index=False)
```

```python
sample_context = squad1_translated.loc[1, 'context_ta']
sample_answer_text = squad1_translated.loc[1, 'answer_text_ta']
sample_answer_start = int(squad1_translated.loc[1, 'answer_start_ta'])
sample_context[sample_answer_start:sample_answer_start+20], sample_answer_text
```

```python
formatted = squad1_translated[squad1_translated['error_ta']=='success'].reset_index(drop=True)
formatted.drop(['title', 'context', 'question', 'answer', 'answer_start_char_idx', 'answer_end_char_idx', 'answer_start_word_idx', 'answer_end_word_idx', 'error_ta'], axis=1, inplace=True)
formatted['answer_start_ta']=formatted['answer_start_ta'].astype(int)
formatted['language']='tamil'
formatted.columns = formatted.columns.str.replace(r'_ta$', '')
formatted
```

```python
formatted.to_csv('squad1.2_ta_formatted.csv', index=False)
```
