```python
import pandas as pd
from tqdm.notebook import tqdm
import json
```

```python
df = pd.read_csv('../../input/translate-cache/squad2_translated_0905.csv')
df = df[(df['answer'].notna())|(df['answer_start_char_idx']==-1)].reset_index(drop=True)
df[df.select_dtypes(include='int64').columns] = df.select_dtypes(include='int64').astype('int32')
df[df.select_dtypes(include='float64').columns] = df.select_dtypes(include='float64').astype('float32')
df
```

```python
df2check = df[df['answer_start_ta'].notnull()]
df2check = df2check[df2check['answer_start_ta']!=-1].reset_index(drop=True)
df2check
```

```python
# sanity check
for i in range(len(df2check)):
    context = df2check.loc[i, 'context_ta']
    answer = df2check.loc[i, 'answer_text_ta']
    start = int(df2check.loc[i, 'answer_start_ta'])
    assert context[start: start+len(answer)]==answer
```

```python
data = {'id': [], 'context': [], 'question': [], 'answers': []}
for i in tqdm(range(len(df))):
    data['id'].append(i)
    status = df.loc[i, 'error_ta']
    if status == 'success':
        data['context'].append(df.loc[i, 'context_ta'])
        data['question'].append(df.loc[i, 'question_ta'])
        answer_text_ta = df.loc[i, 'answer_text_ta']
        if pd.isnull(answer_text_ta):
            data['answers'].append({'text': [], 'answer_start': []})
        else:
            answer_start_ta = int(df.loc[i, 'answer_start_ta'])
            assert answer_start_ta != -1 and not pd.isnull(answer_start_ta)
            data['answers'].append({'text': [answer_text_ta], 'answer_start': [int(answer_start_ta)]})
    else:
        data['context'].append(df.loc[i, 'context'])
        data['question'].append(df.loc[i, 'question'])
        data['answers'].append({'text': [df.loc[i, 'answer']], 'answer_start': [int(df.loc[i, 'answer_start_char_idx'])]})
```

```python
for k, v in data.items():
    print(k, len(v))
    print(v[:5])
```

```python
# last sanity check
for i in tqdm(range(len(data['id']))):
    if len(data['answers'][i]['text']) == 0: continue
    context = data['context'][i]
    question = data['question'][i]
    answer_text = data['answers'][i]['text'][0]
    answer_start = data['answers'][i]['answer_start'][0]
    assert context[answer_start:answer_start+len(answer_text)]==answer_text
```

```python
with open('squad2_enta.json', 'w') as f:
    json.dump(data, f, indent=4)
```

```python
with open('squad2_enta.json', 'r') as f:
    data = json.load(f)
for k, v in data.items():
    print(k, len(v))
    print(v[:5])
```

```python
from sklearn.model_selection import train_test_split
df['if_success'] = df['error_ta']=='success'
df_train, df_valid = train_test_split(df, test_size=0.2, random_state=42, stratify=df['if_success'])
df_train = df_train.reset_index(drop=True)
df_valid = df_valid.reset_index(drop=True)
df_train
```

```python
def to_dict(df):
    data = {'id': [], 'context': [], 'question': [], 'answers': []}
    for i in tqdm(range(len(df))):
        data['id'].append(i)
        status = df.loc[i, 'error_ta']
        if status == 'success':
            data['context'].append(df.loc[i, 'context_ta'])
            data['question'].append(df.loc[i, 'question_ta'])
            answer_text_ta = df.loc[i, 'answer_text_ta']
            if pd.isnull(answer_text_ta):
                data['answers'].append({'text': [], 'answer_start': []})
            else:
                answer_start_ta = int(df.loc[i, 'answer_start_ta'])
                assert answer_start_ta != -1 and not pd.isnull(answer_start_ta)
                data['answers'].append({'text': [answer_text_ta], 'answer_start': [int(answer_start_ta)]})
        else:
            data['context'].append(df.loc[i, 'context'])
            data['question'].append(df.loc[i, 'question'])
            data['answers'].append({'text': [df.loc[i, 'answer']], 'answer_start': [int(df.loc[i, 'answer_start_char_idx'])]})
    return data
train = to_dict(df_train)
valid = to_dict(df_valid)
```

```python
with open('squad2_enta_train.json', 'w') as f:
    json.dump(train, f, indent=4)
with open('squad2_enta_valid.json', 'w') as f:
    json.dump(valid, f, indent=4)
```
